```python
# imports 
import pandas as pd
import numpy as np
import env
```

```python
# need to make a funciton to aquire the url connection
def get_db_url(db, user = env.user, host = env.host, password = env.password):
    '''
    This function will take in a databse and set enviromental credentials and
    return a datbase connection link. 
    '''
    return f'mysql+pymysql://{user}:{password}@{host}/{db}'
```

```python
url = get_db_url('zillow')
#url
```

```python
# need to make query that pulls all the information that is requested
sql_query = '''
select 
    propertylandusetypeid,
    propertylandusedesc,
    bedroomcnt,
    bathroomcnt,
    calculatedfinishedsquarefeet,
    taxvaluedollarcnt,
    yearbuilt,
    taxamount,
    fips
from properties_2017
Left join propertylandusetype as pt
    Using (propertylandusetypeid)
where pt.propertylandusedesc like 'Single%%'
    '''
```

```python
# make a function to get the data using connection url
def new_zillow_data():
    '''
    This function will use a set sql query and return a zillow data frame based on the 
    exercise II in the ds.codeup.com regression: acquistion and prep request
    '''
    
    sql_query = '''
                select
                    id,
                    parcelid,
                    propertylandusetypeid,
                    propertylandusedesc,
                    bedroomcnt,
                    bathroomcnt,
                    calculatedfinishedsquarefeet,
                    taxvaluedollarcnt,
                    yearbuilt,
                    taxamount,
                    fips
                from properties_2017
                Left join propertylandusetype as pt
                    Using (propertylandusetypeid)
                where pt.propertylandusedesc like 'Single%%'
                '''
    df = pd.read_sql(sql_query, get_db_url('zillow'))
    
    return df
```
